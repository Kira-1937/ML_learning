```python
# Feature Engineering
df['date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]

# df['Date']=df['Date_of_Journey'].apply(lambda x:x.split('/').[0])
```

```python
df.columns
```

```python
df['date']=df['date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
# arival time or departue time
df['Arrival_Time']=df['Arrival_Time'].str.split(' ').str[0]



```

```python
df['Arrival_Time'].head(5)
```

```python
df['Arrival_hr']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_min']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

for col in ['Airline', 'Destination', 'Source', 'Additional_Info']:
    transformed = ohe.fit_transform(df[[col]])
    transformed_df = pd.DataFrame(transformed, columns=ohe.get_feature_names_out([col]))
    df = pd.concat([df.drop(col, axis=1), transformed_df], axis=1)
```
